## Como se carega as bases de dados
primeiro e importante enteder que nem todas as bases de dados sao caregas por padrao pelo pandas as veces e necessario usar outras libs no caso de bases em excel e sql 

como estou usando uma base de dados em .csv ele carega por padaro

para caregar uma base de dados usando pandas nso caregamos da seguinte forma:

```python

db1 = 'data/qc_science-and-technology_moz.csv'
db2 = 'data/clearglobal_language_use_moz_admin1.csv'
```

Nos criamos duas cariaves que tem o path(caminho) da localizacao das nocas bases de dados

## como ler essas bases de dados com pandas

```python
pd.read_csv(db1)
```

```python
pd.read_csv(db2)
```

como ves nos exemplos acima lemos duas bases de dados usndo pandas

1 primeiro chamamos o pandas `pd`

2 de seguida nos passamos a fuuncao que le os arquivo csv `.read_csv()`

3 depois dentro da funcao nos passamos a variavel que armazena o path da nossa base de dados `(vraiavel)`
